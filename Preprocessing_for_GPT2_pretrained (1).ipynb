## AMI2020 [CSV] dataset

```
# - - - - - - - - - - - - - - - - - - 
DATASET_PATH = "AMI2020_training_raw.csv" 
# - - - - - - - - - - - - - - - - - -
```

```
import pandas as pd

new_dataset_path = "AMI2020_raw_dataset.txt"

data = pd.read_csv(DATASET_PATH)
data.head()
```

```
import re

def clean(string):
  s = string.strip()
  s = re.sub("@[a-zA-z0-9]*", "", s)                    # remove user names
  s = re.sub("(https:\/\/t.co\/)[a-zA-Z0-9]*", " ", s)  # remove Twitter links
  s = re.sub("\n", " ", s)
  s = re.sub("\s\s(\s*)", " ", s)
  s = s.strip()
  return s
```

```
new_data = []

for row in data[["text", "misogynous", "aggressiveness"]].iterrows():
  text = clean(row[1]["text"])
  mis = row[1]["misogynous"]
  agg = row[1]["aggressiveness"]
  label = max(mis, agg)
  sample = f"[{mis};{agg}] {text}\n"
  if mis == 0 and agg == 0:
    print(mis, agg, text)
  new_data.append(sample)

with open(new_dataset_path, "w") as f:
  f.writelines(new_data)

prompts = []
for row in new_data[:5]:
  p = row[:row.index(";")+3]
  if p not in prompts:
    prompts.append(p)
    print(prompts)
```
